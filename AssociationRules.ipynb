```python
# importing the libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
titanic = pd.read_csv('/Users/furkhanansari/Desktop/Data Science/DataSets/Titanic.csv')
titanic.head()
```

```python
df = pd.get_dummies(titanic)
df
```

```python
# Apriori Algorithm

```

```python
frequent_itemset = apriori(df,min_support=0.1 , use_colnames=True)
frequent_itemset
```

```python
rules = association_rules(frequent_itemset , metric = 'lift' , min_threshold =0.7)
rules
```

```python
# sorting top 20 rules based on height lift vlaues
rules.sort_values('lift' , ascending = False)[0:20]
```

```python
# rules which are having the lift value greater than 1
rules[rules.lift>1]
```
