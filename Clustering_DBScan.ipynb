```python
# importing the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the dataset 
df = pd.read_csv('/Users/furkhanansari/Desktop/Data Science/DataSets/Wholesale customers data.csv');
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns = ['Channel','Region'] , axis=1 , inplace = True)
```

```python
df
```

```python
# converting the df into array so that we can pass it to StandarsScaler to normalze the data
array = df.values
```

```python
array
```

```python
stdScaler = StandardScaler().fit(array)
X = stdScaler.transform(array)
```

```python
# Standardized data (Z-scores)
X
```

```python
# implementing the DBscan
dbscan = DBSCAN(eps=0.8 , min_samples =3)
dbscan.fit(X)
```

```python
# Noise samples are given the label -1(outliers)
dbscan.labels_
```

```python
c1 = pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
c1
```

```python
pd.concat([df,c1],axis=1)
```

```python
df['cluster'] = c1
df
```

```python
df.drop(df[df['cluster'] == -1].index , inplace = True)
```

```python
df = df.reset_index(drop =True)
```
