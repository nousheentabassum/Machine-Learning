```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset('tips')
```

```python
tip
```

```python
#strip plot
sns.stripplot(y = 'tip' , data = tip,jitter = True)
plt.ylabel('tip ₹')
plt.show()
```

```python
#grouping with stripplot()
sns.stripplot(y = 'tip',x ='day',data = tip)
plt.ylabel("tip ₹ ")
plt.xlabel("days")
plt.show()
```

```python
sns.stripplot(y = 'tip',x ='day',data = tip,jitter = False)
plt.ylabel("tip ₹ ")
plt.xlabel("days")
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tip)
plt.xlabel("tips ₹")
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data = tip,hue = 'sex')
plt.xlabel("tip₹")
plt.show()
```

```python
#change orentation
sns.swarmplot(x='tip',y='day',data = tip,hue = 'sex',orient = 'h')
plt.show()
```

```python
# box plot and violin plot
sns.boxplot(x = 'day' , y ='tip' , data = tip)
plt.show()
```

```python
sns.violinplot(x='day',y = 'tip' , data = tip)
# plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day' ,y ='tip' ,data = tip)
plt.subplot(1,2,2)
sns.violinplot(x='day',y ='tip' ,data = tip)
plt.show()
```

```python
# combining plots
sns.boxplot(x='day' ,y ='tip' ,data = tip)
sns.violinplot(x='day',y ='tip' ,data = tip)
plt.show()
```

```python
#joint plot
sns.jointplot(x = 'total_bill' , y = 'tip' , data = tip)
plt.show()
```

```python
#density plot
sns.jointplot(x='tip' , y ='total_bill' , data = tip,kind = 'kde')
```

```python
sns.pairplot(tip)
plt.show()
```

```python
sns.pairplot(tip,hue ='sex')
plt.show()
```
