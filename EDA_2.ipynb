Isolation forest 

```python
from sklearn.ensemble import IsolationForest
```

```python
df = data
```

```python
data_encoded = pd.get_dummies(df)
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state =10 , contamination = .01)
clf.fit(data_encoded)
```

```python
# prediction 
y_pred_outliers = clf.predict(data_encoded)
```

```python
# -1 for outliers and 1 for normal data point
y_pred_outliers
```

```python
data_encoded
```
