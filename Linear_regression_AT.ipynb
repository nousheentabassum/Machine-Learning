```python
import statsmodels.formula.api as smf
model = smf.ols("AT~Waist",data = df).fit()
```

```python
# coefficents 
model.params
```

```python
# t and p values
print( model.tvalues ,'\n', model.pvalues)
```

```python
# r squared vlaues
print(model.rsquared)
```
