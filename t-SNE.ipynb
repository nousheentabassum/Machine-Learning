```python
from pandas import read_csv
import pandas as pd
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
# load the dataset
filename = '/Users/furkhanansari/Desktop/Data Science/DataSets/TSNE_data.csv'
dataframe = pd.read_csv(filename)
```

```python
# splitting the data set for validation 
array = dataframe.values

# seperate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
# TSNE visualizations

data_tsne = TSNE(n_components = 2).fit_transform(X)
cluster.tsneplot(score = data_tsne)
```

```python
# get the list of categories
color_class = dataframe['diagnosis'].to_numpy()
cluster.tsneplot(score = data_tsne , colorlist = color_class , legendpos = 'lower right' , legendanchor =(1.15,1) )
```
