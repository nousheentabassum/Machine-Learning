```python
# vaiable 1 : Smart Thermostats vs. Smart Lights
# variable 2 : Satisfaction(Very Satisfied,Satisfied,Neutral,Unsatisfied,Very Unsatisfied)

# test alpha value
alpha = 0.05

# h0 : satisfaction and item bought are two independent (no associantion)
# h1 : satisfaction and item bought are two not independent (there is associantion)

# caluculate the expencted values 
 # E = (row_total * col_total)/ Grand total
```

```python
import numpy as np
import scipy.stats as stats
```

```python
observed = np.array([
    [50,70],
    [80,100],
    [60,90],
    [30,50],
    [20,50]
])
```

```python
row_total = observed.sum(axis=1)
print(row_total)
```

```python
col_total = observed.sum(axis =0)
print(col_total)
```

```python
grand_total = observed.sum()
print(grand_total)
```

```python
# find the expected values
expected = np.outer(row_total , col_total) / grand_total
print(expected)
```

```python
# chi sqauare statictic
```

```python
chi_square = ((observed - expected)**2/expected).sum()
```

```python
print(chi_square)
```

```python
# dergees of freedom 
dof = (observed.shape[0]-1) * (observed.shape[1]-1)
print(dof)
```

```python
# p value 
p_value = 1 - stats.chi2.cdf(chi_square , dof)
print(p_value)
```

```python
if p_value > alpha :
    print( "fail to reject h0 : ther is no association between them")
else:
    print("fail to reject h1 : ther is association between them")
```

```python
# method 2 shotcut
chi2 , p_value2 , dof2,expected2 = stats.chi2_contingency(observed)
```

```python
print( chi2)
print( p_value2)
print( dof2)
print( expected2)
# same as the above 
```

```python
if p_value2 > alpha :
    print( "fail to reject h0 : ther is no association between them")
else:
    print("reject h0 : ther is association between them")
```

```python
# comparing it with the critical value 
critical_value = stats.chi2.ppf(1 - alpha, dof)
print(critical_value)  # ≈ 9.4877 for dof=4, alpha=0.05
```

```python
if chi_square > critical_value :
    print('reject h0')
else:
    print('fail to reject h0')
```
