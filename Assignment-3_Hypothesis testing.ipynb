```python
import pandas as pd 
import numpy as np
```

```python
# hypothesis testing
# H0 : there is no increse in the cost U = 4000
# H1 : there is increse in the cost u > 4000
x = 3050
mean = 4000
std = 125
n = 25
alfa = 0.05
zvalue = (x - mean )/ (std / np.sqrt(n))
print('Z value : ',zvalue)
```

```python
from scipy import stats
critical_value = stats.norm.ppf(1-alfa)
print( 'Critical Value : ' , critical_value)
```

```python
pvalue =1-stats.norm.cdf(zvalue)
print(pvalue)
```

```python
# fail to reject null hypothesis as 
# the pvalue > alfa
```
