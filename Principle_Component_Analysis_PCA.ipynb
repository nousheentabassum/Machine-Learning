```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
unvi = pd.read_csv('/Users/furkhanansari/Desktop/Data Science/DataSets/Universities.csv')
```

```python
unvi.describe()
unvi.head()
```

```python
# considering only the numerical data
unvi.data = unvi.iloc[:,1:]
unvi.data.head()
UNI = unvi.data.values
UNI
```

```python
# Normalizing the numerical values 
uni_normal = scale(UNI)
```

```python
# Normalized or z-scored
uni_normal
```

```python
# fitting the data to get the pca vaules 
pca = PCA()
pca_values = pca.fit_transform(uni_normal)
pca_values

```

```python
pca = PCA(n_components =6)
pca_values = pca.fit_transform(uni_normal)
```

```python
# the amount of variance that each pca explains is
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# variance plot for PCA coomponents obtained
plt.plot(var1,color='red')
plt.title('cumm varainace')
plt.show()
```
