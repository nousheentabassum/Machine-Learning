```python
# finding out the Confidence Intervals 
from scipy import stats
import pandas as pd
import numpy as np
```

```python
data = pd.Series([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
data
```

```python
# a. Build 99% Confidence Interval Using Sample Standard Deviation
data_CI = stats.t.interval(
    0.99,
    df=len(data)-1,
    loc = data.mean(),
    scale = stats.sem(data)
)
print("Gain at 95% confidence intervals data : " , np.round(data_CI,4))
```

```python
# b. Build 99% Confidence Interval Using Known Population Standard Deviation
std = 0.2
scaleis = std / np.sqrt(len(data))

data_CI_std = stats.norm.interval(
    0.99,
    loc=data.mean(),
    scale = scaleis
)
```

```python
print("Gain at 95% confidence intervals data with std: " , np.round(data_CI_std,4))
```
