```python
# Ex. An outbreak of Salmonella-related illness was attributed to ice cream produced at 
# a certain factory. Scientists measured the level of Salmonella in 9 randomly sampled batches 
# of ice cream. The levels (in MPN/g) were
# 0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418
# Is there evidence that the mean level of Salmonella in the ice cream is greater than 0.3 MPN/g?
# Let us try in Python
```

```python
data = pd.Series([0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418])
```

```python
data
```

```python
# it give the t-value and then pvalue this code one line code for hypo_test
# this is for one smaple t-test code
stats.ttest_1samp(data,0.3)
```

```python
# this is the two tailed test p values so divide by 2 to get final p-value
p = stats.ttest_1samp(data,0.3)[1]
p_value = p/2
```

```python
print(p_value)
```

```python
#data creation with the random values with given n  and p1
n1 = 247
p1 = .37

n2 = 308
p2 = .39
```

```python
population1 = np.random.binomial(1,p1,n1)
population2 = np.random.binomial(1,p2,n2)
```

```python
population1.mean()
```

```python
population2.mean()
```

```python
# for proportion test we use statsmodels lib here we compare the porportions
import statsmodels.api as sm
sm.stats.ttest_ind(population1,population2)
# t-vlaue , p-value , df
```
