```python
# calculating interqurtile range for Net Sales Value
q1 = sale['Net Sales Value'].quantile(0.25)
q2 = sale['Net Sales Value'].quantile(0.75)
IQR = q2-q1
print(f"Interqurtile Range (IQR) : {round(IQR,2)}")
```

```python
# getting a outliers into list

# defineing lower and upper boundries for outliers
l_bound = q1 - 1.5 * IQR
u_bound = q2 + 1.5 * IQR

# filter the dataframe for values outside these bonds

outliers_df = sale[(sale['Net Sales Value'] < l_bound) |(sale['Net Sales Value'] > u_bound)]
# beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
# outliers_df = outliers_df.set_index(outliers_df['Net Sales Value'])
outliers_df = outliers_df.sort_values(by = 'Net Sales Value')
outliers_df
```

```python
# standardization of numerical variables

```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Amount', 'Net Sales Value']

sale[[col + '_scaled' for col in num_cols]] = scaler.fit_transform(sale[num_cols])

sale.head()
```

```python
print("Mean:", sale['Volume_scaled'].mean())
print("Std Dev:", sale['Volume_scaled'].std())
```

```python
# Conversion of Categorical Data into Dummy Variables

```

```python
# Select the categorical columns you want to encode
categorical_cols = ['Day', 'City', 'BU', 'Brand']

# Perform One-Hot Encoding (Creating Dummy Variables)
# drop_first=True prevents multi-collinearity (dummy variable trap) by dropping one redundant column per category
sale_encoded = pd.get_dummies(sale, columns=categorical_cols, drop_first=True)

# Preview your newly expanded dataframe
sale_encoded.head()
```
